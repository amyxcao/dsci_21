### Hello, we are using the User Knowledge dataset. It is about the students' knowledge status about the subject of Electrical DC Machines.


```r
## We loaded the necessary libraries

library(tidyverse)
library(ggplot2)
library(readxl)
```

### We have loaded the data onto a repo on [GitHub](https://github.com/amyxcao/dsci_21/blob/main/Project.ipynb)


```r
## We read the data, it was a excel file with a few pages. TA advised us to convert it into a CSV file and read like. 
## This is only the training data.

data_set <- read_csv("https://raw.githubusercontent.com/amyxcao/dsci_21/main/Copy%20of%20Data_User_Modeling_Dataset_Hamdi%20Tolga%20KAHRAMAN.csv")
data_set
```

## Here are the titles:

- STG (The degree of study time for goal object materials): (input value)
- SCG (The degree of repetition number of user for goal object materials): (input value)
- STR (The degree of study time of user for related objects with goal object): (input value)
- LPR (The exam performance of user for related objects with goal object): (input value)
- PEG (The exam performance of user for goal objects): (input value)
- UNS (The knowledge level of user): (target value)


```r
# The 7th row onwards were instructions, so we had to remove that from our dataframe. 

data_set <- data_set |> 
select(STG:UNS)

data_set
```

```r
# We created a few plots looking at STG and LPR vs STG. Basically, study time, repetition and performance. 
# We looked for data that had attributes which made it easier to classify. 
```

```r
#This were the best tables, from this visualisation we could see that exam performance was classifiable by userknowledge level. 
# More than Study time or repetition of study material

# Plot STG (study time) against PEG (exam marks) with color by UNS 
ggplot(data_set, aes(x = PEG, y = STG, color = UNS)) +
  geom_point() +
  labs(x = "The exam performance of user for goal objects(PEG)", y = "The degree of study time for goal object materails (STG)", title = "STG against PEG with Color by UNS", color = "The knowledge level of user(UNS)") 



# Plot SCG (repretition) against PEG (exam marks) with color by UNS
ggplot(data_set, aes(x = PEG, y = SCG, color = UNS)) +
  geom_point() +
  labs(x = "The exam performance of user for goal objects(PEG)", y = "The degree of repetition number of user for goal object materails(SCG)", title = "SCG against PEG with Color by UNS", color = "The knowledge level of user(UNS)")

```

## Hence, the variable we are interested in looking at are:
- STG (The degree of study time for goal object materials)
- SCG (The degree of repetition number of user for goal object materails)
- PEG (The exam performance of user for goal objects)
- UNS (The knowledge level of user)

## So, lets make a summary table for this information 

```r
# Selecting only the columns STG, PEG, and UNS
selected_data <- select(data_set, STG, SCG, PEG, UNS)

selected_data

# Summary table for this info
```

```r
summarized_data <- selected_data |>
  group_by(UNS) |>
  summarize(count = n(),
            mean_STG = mean(STG),
            mean_SCG = mean(SCG))

                  
summarized_data
```

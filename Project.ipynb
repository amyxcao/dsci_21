```r
library(tidyverse)
library(ggplot2)
```

```r
library(readxl)
```

```r
data_set <- read_csv("https://raw.githubusercontent.com/amyxcao/dsci_21/main/Copy%20of%20Data_User_Modeling_Dataset_Hamdi%20Tolga%20KAHRAMAN.csv")
data_set
```

```r

data_set <- data_set |> 
select(STG:UNS)

data_set
```

```r
plot1 <- data_set |>
ggplot(aes(x = LPR, y = STG, color = UNS)) +
  geom_point() +
  labs(x = "LPR", y = "STG", title = "STG vs LPR with Color by UNS")

plot1
```

```r
plot2 <- data_set |>
ggplot(aes(x = PEG, y = STG, color = UNS)) +
  geom_point() +
  labs(x = "PEG", y = "STG", title = "STG vs LPR with Color by UNS")

plot2
```
